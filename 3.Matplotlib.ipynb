- простые графики

```python
# имитация данных
# массив (вектор-строка) из тысячи значений
# линейной последовательности от 0 до 10
x = np.linspace(0,10, 1000)
x
```

```python
# данные для оси у
# на базе функции sin
y = np.sin(x)
```

```python
# рисуем график
plt.plot(x,y)
plt.show()
```

```python
# совмещение двух графиков в одном окне
plt.plot(x,y)
plt.plot(x,np.cos(x))
plt.show()
```

```python
# интерактивный режим
# %matplotlib notebook

# метки и легенды графиков
plt.plot(x,y, label='sin')
plt.plot(x,np.cos(x), label='cos')
plt.legend()


# напишем заголовок графика
plt.title("Синус и косинус от х")
# метки (подписи осей)
plt.xlabel("x")
plt.ylabel("sin(x) & cos(x)")


plt.show()
```

```python
# цвет и вид линий

# можно задавать разными способами
plt.plot(x, np.sin(x), color='blue') # словарь цветов надо смотреть в документации
plt.plot(x, np.sin(x+1), color='0.75') # rgbcmyk по его перечню букв k = black градации серого
plt.plot(x, np.sin(x+2), color='g')
plt.plot(x, np.sin(x+3), color=(1.0, 0.5, 0.2)) # RGB-кортеж, значения от 0 до 1
plt.show()
```

```python
# вид линий
plt.plot(x, x+3, linestyle='--') # штриховая линия
plt.plot(x, x+2, linestyle=':') # 
plt.plot(x, x+1, linestyle='-.')
plt.show()
```

```python
# объединение кодов вида и цвета линий
plt.plot(x, x+3, '--g') # штриховая зеленая
plt.plot(x, x+2, ':k') # пунктирная черная
plt.plot(x, x+1, '-.m') # штрихпунктирная 
plt.show()
```
