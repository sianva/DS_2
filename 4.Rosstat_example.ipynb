# Анализ данных из Росстата

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#  загрузка данных
stat_data = pd.read_excel('https://rosstat.gov.ru/storage/mediabank/cen-god.xls', skiprows=2, decimal=',').dropna()
stat_data.head(10)
```

```python
stat_data.info()
```

```python
stat_data.columns
```

```python
stat_data.columns = stat_data.columns.astype('str')
stat_data.columns
```

```python
# транспонирование матрицы (таблицы)
stat_data = stat_data.T
stat_data.head()
```

```python
# в качестве ключей столбцов устанавливаем значения столбцов
stat_data.columns = stat_data.loc[' ПОКАЗАТЕЛИ ', :]
stat_data.head()
```

```python
stat_data.iloc[:, :10]
```

```python
# срезаем из таблицы данные до 1998 г, т.к. там был присутствуют аномальные значения
stat_data = stat_data.loc['1998':, :]
stat_data.head()
```

```python
stat_data.iloc[:, 60:]
```

```python
# символ "-" заменяем на 0
stat_data = stat_data.replace('-', 0.0)
stat_data.iloc[:, 60:]
```

```python
# проверка значений 
stat_data.values
```

```python
# конвертирование строковых значений в числовой тип (float)
stat_data = stat_data.astype(np.float32)
stat_data.values
```

```python
stat_data.columns
```

```python
# удаление лишних символов - пробела (" ") из ключей столбцов
stat_data.columns = stat_data.columns.str.strip(" ")
stat_data.columns
```

```python
# 'очищенный' DataFrame можно сохранить в виде нового ексель файла
stat_data.to_excel('prices_stat_data_1998-2019.xlsx')
```

```python
stat_data['Говядина (кроме бескостного мяса)']
```

```python
# в сколько раз изменилась средняя цена на определенную категорию
stat_data.loc['2019', 'Говядина (кроме бескостного мяса)']/stat_data.loc['1998', 'Говядина (кроме бескостного мяса)']
```

```python
stat_data.loc['2019', 'Куры охлажденные и мороженые1)']/stat_data.loc['1998', 'Куры охлажденные и мороженые1)']
```

```python
# извлечение интересующих данных
prices_1 = stat_data['Куры охлажденные и мороженые1)']
prices_2 = stat_data['Мыло туалетное, за 100 г']
prices_3 = stat_data['Кирпич красный, за 1000 шт.']
```

```python
# извлекаем года для временной шкалы, попутно (на лету) конвертируя в целочисленный тип
years = stat_data.index.astype(int)
years
```

```python
# визуализация
plt.plot(years, prices_1, label='Куры охлажденные и мороженые1)')
plt.plot(years, prices_2, label='Мыло туалетное, за 100 г')
plt.plot(years, prices_3, label='Кирпич красный, за 1000 шт.')
plt.legend()
plt.title('Динамика средних цен с 1998 по 2019 гг')
plt.xlabel('годы')
plt.ylabel('средние цены (руб)')
plt.show()
```
